## Model regresi logistik biner yang akan digunakan  sebagai berikut 
$$
\text{logit}(p)  =  \log\left(\frac{p}{1-p}\right) =\beta_0 + \beta_1 X_1 + \beta_2 X_2 + \beta_3 X_3  + \beta_4 X_4 
$$



```python
import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Pastikan df adalah DataFrame yang sudah dipersiapkan sebelumnya
X = df[["X2_JK_KRT", "X4_ijz_KRT", "X5_bekerja_KRT", "X17_motor"]]  # Fitur
y = df["X16_layanan_keuangan"]  # Target

# Membagi data menjadi data pelatihan dan pengujian
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1, random_state=3692)

# Melakukan standardisasi fitur
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)  # Mengubah menjadi numpy array 2D
X_test = scaler.transform(X_test)  # Mengubah menjadi numpy array 2D

# Menambahkan konstanta (intercept) pada fitur
X_train_with_intercept = sm.add_constant(X_train)

# Membuat dan melatih model regresi logistik menggunakan statsmodels
logit_model = sm.MNLogit(y_train, X_train_with_intercept)  # MNLogit untuk regresi logistik multinomial
result = logit_model.fit()

# Menampilkan hasil estimasi parameter (koefisien dan p-value)
print(result.summary())
```

Berdasarkan hasil pengelolan dihasilkan model regresi Logistik Biner untuk mengetahui pengaruh akses layananan keuangan sebagai berikut 

$$
\text{logit}(p)  = -1.2230 + 0.1515 X_1 -0.1407  X_2 +   0.1820 X_3  + 0.1650   X_4 
$$

$$
